```python
import yfinance as yf
import pandas as pd
```

```python
stocks = [
    "RELIANCE.NS",
    "INFY.NS",
    "TCS.NS",
    "HDFCBANK.NS",
    "SUNPHARMA.NS",
    "AXISBANK.NS",
    "HINDUNILVR.NS"
]
```

```python
import time

data = pd.DataFrame()

for s in stocks:
    temp = yf.download(s, start="2020-01-01", end="2025-01-01")
    temp = temp["Close"]
    data[s] = temp
    time.sleep(2)

data.head()
```

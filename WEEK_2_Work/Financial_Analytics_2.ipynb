```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('stock_data.csv')
```

```python
data
```

```python
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace = True)
data.head()
```

```python
## returns in %

returns = data.pct_change()
returns.head()
```

```python
## remove NaN

returns = returns.dropna()
returns.head()
```

```python
# To find Avg returns

returns.mean()
```

### Insights
 SUNFARMA -> highest returns   -> 0.13 % pre day returns
 HINDUNILVR -> lowest returns

```python
## Risk (Volatility)

returns.std()
```

### From Return and Risk Insights ->
    SUNFARMA -> HIGH - MEDIUM -> BEST
    INFY -> GOOD - MEDIUM -> BALANCED
    AXISBANK -> MEDUIM - HIGH -> RISKY
    HINDUNILVR -> LOW - LOW -> SAFE

```python
# To find correlation

returns.corr()
```

### Insights

IT stocks like TCS and INFY -> high corr
SUNFARMA and HINDUNILVR -> lower corr

```python
import seaborn as sns

sns.heatmap(returns.corr(), annot = True)
plt.show()
```
